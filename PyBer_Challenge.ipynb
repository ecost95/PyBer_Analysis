### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
city_data_df
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head(10)
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
# use the groupby() function to create a Series of data that has the type of city as the index, 
# then apply the count() method to the "ride_id" column.

all_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
all_ride_count
```

```python
# 2. Get the total drivers for each city type

all_driver_count = city_data_df.groupby(["type"]).sum()["driver_count"] 
all_driver_count
```

```python
#  3. Get the total amount of fares for each city type
all_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
all_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_ride = all_fares / all_ride_count
average_fare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_driver = all_fares / all_driver_count
average_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
frame = { 'Total Rides': all_ride_count, 'Total Drivers': all_driver_count, 'Total Fares': all_fares, 'Average Fare per Ride': average_fare_ride, 'Average Fare per Driver': average_fare_driver}
pyber_summary_df = pd.DataFrame(frame)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df.dtypes
```

```python
round(pyber_summary_df['Average Fare per Ride'], 2)
```

```python
#  8. Format the columns.

pyber_summary_df.loc[:, 'Average Fare per Ride'] = round(pyber_summary_df['Average Fare per Ride'], 2)
pyber_summary_df.loc[:, 'Average Fare per Driver'] = round(pyber_summary_df['Average Fare per Driver'], 2)

pyber_summary_df['Total Drivers'] = pyber_summary_df.apply(lambda x: "{:,}".format(x['Total Drivers']), axis=1)
pyber_summary_df['Total Rides'] = pyber_summary_df.apply(lambda x: "{:,}".format(x['Total Rides']), axis=1)
pyber_summary_df['Total Fares'] = pyber_summary_df.apply(lambda x: "{:,}".format(x['Total Fares']), axis=1)


pyber_summary_df
#pyber_summary_df
#pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map(lambda x: f"${0:,.2f}")
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df.head(10)
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

fares_by_date = pyber_data_df.groupby(["type", "date"]).sum()["fare"]

fares_by_date_df = pd.DataFrame(fares_by_date)
fares_by_date_df.head()
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_by_date_df = fares_by_date_df.reset_index()

fares_by_date_df.head()
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fares_by_date_pivot = fares_by_date_df.pivot(index = "date", columns="type", values="fare")
fares_by_date_pivot.head(10)
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.


fares_by_date_df = fares_by_date_pivot.loc[(fares_by_date_pivot.index > "2019-01-01") & (fares_by_date_pivot.index < "2019-04-28")]
fares_by_date_df.tail()
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fares_by_date_df.index = pd.to_datetime(fares_by_date_df.index)
fares_by_date_df
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fares_by_date_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fares_by_date_week = fares_by_date_df.resample("W").sum()
fares_by_date_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import seaborn as sns
# Use the graph style fivethirtyeight.

style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(18,6))


# # Plotting each occupation category
ax1 = sns.lineplot(x=fares_by_date_week.index, y=fares_by_date_week['Rural'].tolist(), color='#0F95D7', lw=2.5, label="Rural")
ax2 = sns.lineplot(x=fares_by_date_week.index, y=fares_by_date_week['Suburban'].tolist(), color='#FF2700', lw=2.5, label="Suburban")
ax3 = sns.lineplot(x=fares_by_date_week.index, y=fares_by_date_week['Urban'], color='yellow', lw=2.5, label="Urban")
ax.set_title('Total Fare by City Type',fontsize=20)
ax.set_ylabel('Fare ($USD)',fontsize=14)
ax.set_xlabel('Date',fontsize=14)
ax.legend()

# Save the figure.
plt.savefig("analysis/PyBer_fare_summary.png")
```
